```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
from scipy import stats
```

```python
# Given data 
mean = 5 
std_dev = 1.5
```

```python
# Function to calculate the cumulative distribution function (CDF) 

def calculate_proportion(lower_bound, upper_bound=None): 
    if upper_bound is None: 
        return norm.cdf(lower_bound, mean, std_dev) 
    return norm.cdf(upper_bound, mean, std_dev) - norm.cdf(lower_bound, mean, std_dev)
```

```python
# 2.1 Proportion of gunny bags with breaking strength < 3.17 kg/cm^2 
prop_less_than_3_17 = calculate_proportion(3.17) 
print(f"Proportion < 3.17 kg/cm^2: {prop_less_than_3_17:.4f}") 
```

```python
# 2.2 Proportion of gunny bags with breaking strength >= 3.6 kg/cm^2 
prop_greater_than_3_6 = 1 - calculate_proportion(3.6) 
print(f"Proportion >= 3.6 kg/cm^2: {prop_greater_than_3_6:.4f}") 
```

```python
# 2.3 Proportion of gunny bags with breaking strength between 5 and 5.5 kg/cm^2 
prop_between_5_and_5_5 = calculate_proportion(5, 5.5) 
print(f"Proportion between 5 and 5.5 kg/cm^2: {prop_between_5_and_5_5:.4f}")
```

```python

# 2.4 Proportion of gunny bags with breaking strength NOT between 3 and 7.5 kg/cm^2 
prop_not_between_3_and_7_5 = calculate_proportion(3) + (1 - calculate_proportion(7.5)) 
print(f"Proportion NOT between 3 and 7.5 kg/cm^2: {prop_not_between_3_and_7_5:.4f}")
```

```python
# Visual representation 
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000) 
y = norm.pdf(x, mean, std_dev)
```

```python
            
plt.figure(figsize=(12, 6))
plt.plot(x, y)
plt.fill_between(x, 0, y, where=(x < 3.17), color='r')
plt.axvline(x=3.17,color='r')
plt.title("Proportion < 3.17 kg/cm^2") 
plt.xlabel("Breaking Strength (kg/cm^2)") 
plt.ylabel("Density") 
plt.legend()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(x, y, label='Normal Distribution') 
plt.fill_between(x, 0, y, where=(x >= 3.6), color='r')
plt.axvline(x=3.6,color='r')
plt.title("Proportion >= 3.6 kg/cm^2") 
plt.xlabel("Breaking Strength (kg/cm^2)") 
plt.ylabel("Density")
```

```python
plt.figure(figsize=(12, 6))
plt.plot(x, y) 
plt.fill_between(x, 0, y, where=((x >= 5) & (x <= 5.5)), color='r')
plt.axvline(x=5,color='r')
plt.axvline(x=5.5,color='r')
plt.title("Proportion between 5 and 5.5 kg/cm^2") 
plt.xlabel("Breaking Strength (kg/cm^2)") 
plt.ylabel("Density")
```

```python
plt.figure(figsize=(12, 6))
plt.plot(x, y) 
plt.fill_between(x, 0, y, where=(x < 3) | (x > 7.5), color='r')
plt.axvline(x=3,color='r')
plt.axvline(x=7.5,color='r')
plt.title("Proportion NOT between 3 and 7.5 kg/cm^2") 
plt.xlabel("Breaking Strength (kg/cm^2)") 
plt.ylabel("Density")
```

Problem 3

Zingaro stone printing is a company that specializes in printing images or patterns on polished or unpolished stones. However, for the optimum level of printing of the image, the stone surface has to have a Brinell's hardness index of at least 150. Recently, Zingaro has received a batch of polished and unpolished stones from its clients. Use the data provided to answer the following (assuming a 5% significance level);

3.1 Zingaro has reason to believe that the unpolished stones may not be suitable for printing. Do you think Zingaro is justified in thinking so?

3.2 Is the mean hardness of the polished and unpolished stones the same?

e hypot

```python
Zingaro = pd.read_csv('Zingaro_Company.csv')
Zingaro.head()
```

```python
Zingaro.tail()
```

```python
# 3.1 One-sample t-test for unpolished stones

t_stat, p_value = stats.ttest_1samp(Zingaro['Unpolished '], popmean = 150, alternative='less')

print('The p-value is ', p_value)
```

Let's write the null and alternative hypothesis

Let μ be the mean delivery time of the Zingaro data.

test of null hypothesis:

    Null Hypothesis (H₀): μ ≥ 150 (The mean hardness is at least 150)

test of alternate hypothesis:

    Alternative Hypothesis (Hₐ): μ < 150 (The mean hardness is less than 150)

```python
# Decision rule at 5% significance level 

alpha = 0.05

if p_value < alpha: 
    
    print("Reject the null hypothesis: The unpolished stones are not suitable for printing.") 

else: 
    
    print("Fail to reject the null hypothesis: The unpolished stones are suitable for printing.")
```

```python
# Plot histograms 
plt.figure(figsize=(12, 6)) 

plt.hist(Zingaro['Unpolished '], bins=10, alpha=0.5, label='Unpolished') 
plt.axvline(x=150, color='r', label='Population Mean') 
plt.title('Hardness Distribution of Unpolished Stones') 
plt.xlabel('Hardness Index') 
plt.ylabel('Frequency') 
plt.legend()
```

#3.2 Is the mean hardness of the Treated and Polished and unpolished stones the same?

To compare the mean hardness between Treated and Polished and unpolished stones, 

Conducting a two-sample t-test.

hypothesis:

Null Hypothesis (H₀): μ₁ = μ₂ (The mean hardness of polished and unpolished stones is the same)

Alternative Hypothesis (Hₐ): μ₁ ≠ μ₂ (The mean hardness of Treated and Polished and unpolished stones is different)

```python
t_stat, p_value = stats.ttest_ind(Zingaro['Treated and Polished'], Zingaro['Unpolished '], equal_var=False)

print('The p-value is ', p_value)
```

```python
# Decision rule at 5% significance level 

alpha = 0.05

if p_value < alpha: 
    
    print("Reject the null hypothesis: The mean hardness of polished and unpolished stones is different.") 

else: 
    
    print("Fail to reject the null hypothesis: The mean hardness of polished and unpolished stones is the same.")
```

```python
import statsmodels.api as sm 
from statsmodels.formula.api import ols

def perform_anova(data, alloy_type): 
    print(f"\nANOVA results for Alloy {alloy_type}:\n")
 
    # Two-way ANOVA for interaction effect between dentist and method 
    model_interaction = ols('Response ~ C(Dentist) * C(Method)', data=data).fit() 
    anova_interaction = sm.stats.anova_lm(model_interaction, typ=2) 
    print("Two-way ANOVA for Interaction between Dentist and Method:\n", anova_interaction) 

    # Two-way ANOVA for dentists and methods together 
    model_dentist_method = ols('Response ~ C(Dentist) + C(Method)', data=data).fit() 
    anova_dentist_method = sm.stats.anova_lm(model_dentist_method, typ=2) 
    print("Two-way ANOVA for Dentists and Methods:\n", anova_dentist_method) 
```

```python
 # Perform ANOVA for Alloy 1 
perform_anova(alloy1_df, 1) 

# Perform ANOVA for Alloy 2
perform_anova(alloy2_df, 2)
```
