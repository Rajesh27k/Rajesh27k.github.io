```python
# Convert object to DateTime
df['Date']  = pd.to_datetime(df['Date'])  ## Complete the code to convert Date column from object to datetime
df.head()
```

# Stock Price Analysis

```python
numeric_columns = df.select_dtypes(include=np.number).columns.tolist()  ## Complete the code to get all the numerical columns in the data

for i, stock in enumerate (numeric_columns):
    plt.figure(figsize=(10, 6))
    plt.scatter(df['Date'], df[stock])  ## Complete the code to plot a scatterplot for all the columns against Date
    plt.title(f'{stock} Price Over Time')
    plt.xlabel('Date')
    plt.ylabel('Stock Price')
    plt.grid(True)
    plt.show()
```

## **Inference**

Dish TV has experienced a significant and prolonged decline in stock price, indicating sustained underperformance and high market risk, with little sign of recovery in recent years.

Infosys has shown a strong and consistent upward trend, especially during the pandemic years, reflecting robust fundamentals and investor confidence in the technology sector.

Hindustan Unilever demonstrated impressive growth until 2020, after which it entered a phase of volatility and mild decline, suggesting possible market saturation or macroeconomic pressures affecting consumer demand.

Vodafone Idea has undergone a dramatic collapse in value, with occasional speculative spikes, highlighting its status as a highly volatile and distressed stock with elevated risk.

Cipla has shown steady and accelerating growth, particularly after 2020, positioning it as a relatively low-risk, high-consistency performer in the pharmaceutical sector.

# Returns and Volatility Analysis


## Return Calculation

```python
Return_of_Stocks =  np.log(df.drop(['Date'],axis=1)).diff(axis = 0)

# Set the display options to show all rows and columns
pd.set_option('display.max_rows', None)  # Set to None to display all rows

Return_of_Stocks
```

**Inference**
-


## Average Returns

```python
StockMeans = Return_of_Stocks.mean()  ## Complete the code to get the mean for the returns for all stocks
StockMeans.sort_values()
```

## **Inference**
Cipla has the highest average return at 0.2538% per week, indicating strong and consistent performance over time.

Hindustan Unilever and Infosys follow closely, with average weekly returns of 0.2294% and 0.2180% respectively—both reflecting solid long-term growth potential.

Dish TV and Vodafone Idea have negative average returns (−0.3751% and −0.3932% per week), confirming their poor historical performance and reinforcing their classification as high-risk, underperforming assets.


## Volatility

```python
StockStdDev = Return_of_Stocks.std()  ## Complete the code to get the std. deviation for the returns for all stocks
StockStdDev.sort_values()
```

## **Inference**
* Cipla offers the best risk-return tradeoff high return with moderate volatility.

* Hindustan Unilever is the most stable stock, with the lowest volatility and a solid return—ideal for conservative investors.

* Infosys balances growth and risk well, making it a strong core holding.

* Dish TV and Vodafone Idea have both high volatility and negative returns, making them unattractive from a risk-adjusted perspective.


## Visualizing Returns and Volatility

```python
data = pd.DataFrame({'Mean': StockMeans,
                     'Volatility': StockStdDev})## Complete the code to get a dataframe for mean  and std. deviation for the returns of all stocks
data
```

```python
import seaborn as sns

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111)
plt.scatter(data['Volatility'], data['Mean'])  ## Complete the code to get the plot of returns vs volatility for the returns for all stocks
plt.axhline(y=0,linestyle='--', color = "red")

for index, row in data.iterrows():
    ax.text(row['Volatility'], row['Mean'], index)
plt.xlabel('Volatility')
plt.ylabel('Net returns')
plt.title('Net return vs Volatility for stocks')
plt.grid()
plt.show()
```

## **Inference**
1. Cipla, Hindustan Unilever, and Infosys form a favorable cluster with:

    * Low volatility (around 0.04)

    * Positive net returns (around 0.002 or 0.2% weekly)

    * These stocks offer a strong risk-return profile and are ideal for building a stable, growth-oriented portfolio.

2. Dish TV and Vodafone Idea are positioned in the high-risk, low-return quadrant:

    * High volatility (above 0.08)

    * Negative net returns (around −0.004 or −0.4% weekly)

    * These stocks are underperforming and carry significant downside risk, making them unsuitable for conservative or risk-averse investors.
